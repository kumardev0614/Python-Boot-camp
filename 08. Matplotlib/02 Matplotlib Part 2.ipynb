# Subplots Using Object Oriented Method

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)

y = x ** 2

print('x =', x)
print('y =', y)
```

```python
# This is tupple unpacking. Don't know how?
fig, axes = plt.subplots(nrows=1, ncols=2)

# Here axes is a list of axes objects, which is itterable and indexable.
axes[0].plot(x,y)
axes[0].set_title('axes[0]')

axes[1].plot(y,x)
axes[1].set_title('axes[1]')
```

```python
axes   # axes i holdind 2 objects bc it has 2 subplots. 
```

### More Example

```python
fig, axes = plt.subplots(nrows=2, ncols=3)

axes[0,0].plot(x**2,y)
axes[0,1].plot(x,y)
axes[0,2].plot(x**3, y)

axes[1,0].plot(x, np.exp(y))
axes[1,1].plot(x,y)
axes[1,2].plot(x**3, y)

plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))  # Here dpi=none, and we will not change it.
# size is in inches. Means width= 8 inch and height= 2 inch

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(5,5))  

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

## Subplots with Figure size and DPI

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,6))

ax[0].plot(x,y)

ax[1].plot(y,x)

plt.tight_layout()
```

# axes.legend()

```python
# It is uesd to identify which graph is which graph

myfig = plt.figure()

myax = myfig.add_axes([0,0,1,1])

myax.plot(x, x**2, label='x square')
myax.plot(x, x**3, label='x cube')

myax.legend()

# We can set the location of legend by myax.legend(0),   myax.legend(5),    myax.legend(8)   etc.
```

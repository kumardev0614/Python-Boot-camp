```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 2
y
```

# Functional Method to plot a Graph

```python
plt.plot(x,y)
plt.xlabel('X lable')
plt.ylabel('Y label')
plt.title('TITLE')
```

### Multiplot on same canvas/output

```python
# plot1
# subplot(row,column,plotNumber)

plt.subplot(1,2,1)
plt.plot(x,y,'r')         # 'r' for red

# plot2
plt.subplot(1,2,2)
plt.plot(y,x,'g')         # 'g' for green
```

# Object Oriented Method

```python
# Create a figure object
fig = plt.figure()

# Create axes for this figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])         # add_axes(left, bottom, width, height)

# Noe we can plot on these axes
axes.plot(x,y)
axes.set_xlabel('this is x')
axes.set_ylabel('this is y')
axes.set_title('TITLE')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title('bigger plot')

axes2.plot(y,x)
axes2.set_title('smaller plot')
```

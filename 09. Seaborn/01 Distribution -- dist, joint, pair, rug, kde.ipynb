```python
import seaborn as sns
%matplotlib inline
```

# Load some data

```python
# It is use to plot statistical plots.
# We will use built-in dataset of seaborn library

tips = sns.load_dataset('tips')
```

```python
tips
```

# distplot()

It is use to see the distribution of a single variable(univariate distribution).

```python
sns.distplot(tips['total_bill'])
```

```python
# We can revove this line
sns.distplot(tips['total_bill'], kde=False)
```

```python
# We can increase the number of bins
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

# jointplot()

It is used to see correlation between two variables, and compare them with each other.

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# We have total bill at x axis and tip at y axis.
```

```python
# by default jointplot() shows a scatter plot between variables, but we can change it by argument "kind"
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # regrresion
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

# pairplot()

It creates all possible plots of numerical columns for a dataframe.

```python
sns.pairplot(tips)
```

```python
# we can add hue argument to see categorical data also.
sns.pairplot(tips, hue='sex')
# We can add palette='coolwarm' also to choose color theme/custom colors.
```

# rugplot()

It also used for single variable, need more knowledge to understand rug plots.

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.rugplot(tips['size'])
```

```python
sns.rugplot(tips['tip'])
```

# kdeplot()

```python
sns.kdeplot(tips['total_bill'])
```

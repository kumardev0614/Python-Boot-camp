```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

# barplot()

```python
sns.barplot(x='sex',y='tip', data=tips)
# avrg tip given by males vs avrg tip given by females.
# We can use estimator='np.std'     as aggrigatte function for related columns, default is MEAN.
```

# countplot()

Works as tips['sex'].value_count() function

```python
sns.countplot(x='sex', data=tips)

# here default estimator is count()
# Also y is always set to count.
```

```python
tips['day'].value_counts()
```

```python
sns.countplot(x='day', data=tips)
```

# boxplot()

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# with hue
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

# violinplot()

```python
# it is similer to boxplot but it takes all the datapoints to create the plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# We can add hue also
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# We can see this data side by side to compare using, split=True
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

# stripplot()

It shows data as scatter plot between a categorical and continuous data so it looks like a strip.

```python
sns.stripplot(x='day', y='total_bill', data=tips)  # jitter is by default True.
```

```python
# We can add hue and split also
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')   # we can add split=True also
```

# swormplot()

It is the combination of violin + strip plots.

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

# Combine violin + sworm

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='white')
```

# catplot()/factorplot()

We can create all kind of plots using this plot. By defining the kind='plotname'.

```python
#sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')

# The `factorplot` function has been renamed to `catplot`.
# kind='strip' by default

sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='sex', data=tips, kind='count')
```

# Handling Missing Data

### Imputing Age

```python
# Here we can impute the age as Mean value. But we will try better approach.
# We will impute the age as Mean of Their Passenger Class.
# In order to get the mean we will use boxplot

plt.figure(figsize=(10,8))
sns.boxplot(x='Pclass', y='Age', data=train)

print('From this boxplot, we can see the people who were in 1st calss was higher in age.')
print('3rd class passengesr were mostly young')
print('It is quiet obious, becase it take times to become wealthy to travel in 1st class')
```

```python
def impute_age(age_pclass):
    Age = age_pclass[0]
    Pclass = age_pclass[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age        
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)
```

```python
sns.heatmap(train.isnull())                                    # Train
```

```python
test['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis=1)  # For Test
```

```python
sns.heatmap(test.isnull())                                      # Test
```

### Droping the missing values

```python
# Here in cabin column, There are too much missing values so we will just drop the whole column
train.drop('Cabin', axis=1, inplace=True)
```

```python
test.drop('Cabin', axis=1, inplace=True)
```

```python
sns.heatmap(train.isnull())                                        # Test
```

```python
sns.heatmap(test.isnull())                                           # Test
```

```python
# Ok so there is no any null value in Test dataset, But there is one in Train.
# so will use train.dropna() this will drop all tha rows containing null values.
# Because there is only one row which contains null value so it will only drop single row from Train

train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull())                                                 # Train
```

## Now we don't have any missing value in both Train and Test

# Creating Dummy Variables for Categorical Data

```python
# Here Sex and Embarked columns should be included in Train dataset for our analysis, we cannot just drop them.
# Here our ML algo can't work with string type data. So we will create dummy variabls for the columns which we have to 
# use as inputs. 
```

```python
pd.get_dummies(train['Sex'])
```

```python
# There is a issue here. One column is perfect predictor of another column.
# This problem is known as multicollinearity.

# To avoid this we will drop the forst column

sex = pd.get_dummies(train['Sex'], drop_first=True)
sex
```

```python
# Same for embarked
embarked = pd.get_dummies(train['Embarked'], drop_first=True)
embarked
```

```python
# Now we will concatinate the newly created dummy variables in dataset
train
```

```python
train = pd.concat([train, sex, embarked], axis=1)
# Here we havn't changed the train dataset. We have just concatinated 3 different datasets train, sex, embarked.
# Then saved this new dataset as train. We can change the name from train if we want to.

train
```

### We will do the same for test dataset also

```python
sex = pd.get_dummies(test['Sex'], drop_first=True)
embark = pd.get_dummies(test['Embarked'], drop_first=True)
test = pd.concat([test, sex, embark], axis=1)
```
